```python
lst=["LIMIT_BAL", "SEX","EDUCATION", "MARRIAGE", "AGE", "PAY_0", "PAY_2", "PAY_3", "PAY_4", "PAY_5", "PAY_6", "BILL_AMT1", "BILL_AMT2", "BILL_AMT3", "BILL_AMT4", "BILL_AMT5", "BILL_AMT6", "PAY_AMT1", "PAY_AMT2", "PAY_AMT3", "PAY_AMT4", "PAY_AMT5", "PAY_AMT6"]
lst2=[]
from scipy.stats import pearsonr
for i in lst:
    corr,_ = pearsonr(df["default.payment.next.month"],df[i])
    print(i,'Pearsons correlation: %.3f' % corr)
    if corr<-0.5 or corr>0.5:
        lst2.append(i)
print(lst2)
```

```python
from sklearn.model_selection import train_test_split

#divide x_data and y_data
y_data = df['default.payment.next.month']
x_data=df.drop(['default.payment.next.month','ID'],axis=1)

x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.3, random_state=1)
#random_state=1 gives you different results everytime

print("Number of test samples :", x_test.shape[0])
print("Number of training samples:",x_train.shape[0])
```

```python
############## LOGISTIC REGRESSION

from sklearn.linear_model import LogisticRegression
LR = LogisticRegression(random_state=0,class_weight="balanced").fit(x_train,y_train)
LR
```

```python
yhat1 = LR.predict(x_test)
yhat1
```

```python
from sklearn.metrics import jaccard_score
jaccard_score(y_test, yhat1)
```

```python
from sklearn.metrics import log_loss
log_loss(y_test, yhat1)
```

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix,precision_score, recall_score, f1_score, accuracy_score
confusion_matrix(y_test, yhat1, labels=[1,0])
```

```python
from sklearn.metrics import precision_recall_fscore_support
precision_recall_fscore_support(y_test, yhat1, average='macro')
```

```python
print('Precision: %.3f' % precision_score(y_test, yhat1))
print('Recall: %.3f' % recall_score(y_test, yhat1))
print('Accuracy: %.3f' % accuracy_score(y_test, yhat1))
print('F1 Score: %.3f' % f1_score(y_test, yhat1))
```

```python
####################### KNN

from sklearn.neighbors import KNeighborsClassifier
k = 3
#Train Model and Predict  
neigh = KNeighborsClassifier(n_neighbors = k).fit(x_train,y_train)
neigh
```

```python
knnyhat = neigh.predict(x_test)
knnyhat[0:10]
```

```python
from sklearn.metrics import jaccard_score
jaccard_score(y_test, knnyhat)
```

```python
from sklearn.metrics import log_loss
log_loss(y_test, knnyhat)
```

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, knnyhat, labels=[1,0])
```

```python
from sklearn.metrics import precision_recall_fscore_support
precision_recall_fscore_support(y_test, knnyhat, average='macro')
```

```python
print('Precision: %.3f' % precision_score(y_test, knnyhat))
print('Recall: %.3f' % recall_score(y_test, knnyhat))
print('Accuracy: %.3f' % accuracy_score(y_test, knnyhat))
print('F1 Score: %.3f' % f1_score(y_test, knnyhat))
```

```python
############ DECISION TREE

from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
decision_tree = tree.DecisionTreeClassifier(criterion="entropy", max_depth = 5)
decision_tree = decision_tree.fit(x_train, y_train)
```

```python
ypred = decision_tree.predict(x_test)
score = accuracy_score(y_test, ypred)
print(score)
```

```python
score1 = decision_tree.score(x_test, y_test)
print(score1)
```
